```python
import pandas as pd
import matplotlib.pyplot as plt

# Ganti dengan lokasi file data Anda
data_file = "data_convert.csv"

# Membaca data menjadi dataframe Pandas
df = pd.read_csv(data_file)

# Pilih kolom sensor yang ingin Anda visualisasikan
sensor_column = 'sensor measurement 1'  # Ganti dengan nama kolom sensor yang Anda inginkan

# Membuat scatter plot untuk kolom sensor yang dipilih
plt.figure(figsize=(8, 6))
plt.scatter(df.index, df[sensor_column], s=5, alpha=0.5)
plt.xlabel('Data Point')
plt.ylabel(sensor_column)
plt.title(f"Scatter Plot for {sensor_column}")
plt.grid()
plt.show()
```
